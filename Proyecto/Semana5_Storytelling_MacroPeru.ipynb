## Storytelling

*Storytelling* es una habilidad que combina el análisis de datos con las técnicas narrativas. Se trata de comunicar los resultados de un análisis de manera comprensible y persuasiva para guiar a la audiencia a una conclusión o acción determinada. Algunos criterios a tener consideración: 

**1. Objetivo:** Antes de empezar a contar una historia con datos, es esencial determinar qué mensaje quieres comunicar. ¿Qué acción deseas que tome la audiencia después de ver tus datos?

**2. Conocer a la audiencia:** Saber quién es tu público objetivo te permitirá adaptar el mensaje a sus necesidades y nivel de conocimiento.

**3. Selección de datos:** No todos los datos son igualmente importantes. Es fundamental identificar y presentar solo aquellos datos que respalden directamente tu mensaje.

**4. Visualización:** Una buena visualización de datos (como gráficos, tablas o mapas) puede facilitar enormemente la comprensión del mensaje. Es esencial elegir el tipo de visualización más adecuado según los datos y el mensaje que se quiere transmitir.

**5. Narrativa:** El storytelling implica construir una trama alrededor de los datos. Esto significa establecer un inicio (contexto o problema), desarrollo (análisis o descubrimientos clave) y desenlace (conclusión o llamado a la acción).

**6. Simplicidad:** En general, es mejor mantener las cosas simples y claras. Evitar visualizaciones complejas o abrumadoras y enfocarse en transmitir el mensaje de manera efectiva.

**7. Consistencia:** Es vital mantener una consistencia en los colores, fuentes y estilo general para que la presentación sea armoniosa y fácil de seguir.

**8. Revisión y retroalimentación:** Antes de presentar la historia final, es útil obtener retroalimentación de colegas o miembros de la audiencia objetivo para garantizar que el mensaje sea claro y convincente.

## Principales tipos de gráficos y usos

Los tipos más comunes de gráficos en Seaborn y sus usos son los siguients:

1. Gráfico de barras `(barplot)`:
Uso: Es útil para comparar una cantidad entre un conjunto discreto de ítems.
Ejemplo: Mostrar el promedio de ventas de diferentes productos.


2. Gráfico de dispersión `(scatterplot)`:
Uso: Ideal para visualizar la relación entre dos variables numéricas.
Ejemplo: Comparar el peso y la altura de un grupo de personas.


3. Line plot `(lineplot)` 
Uso: Ideal para visualizar una serie de datos en función del tiempo o una variable secuencial.
Ejemplo: Visualizar las ventas de un producto a lo largo del tiempo.
   
   
4. Histograma `(histplot)`:
Uso: Muestra la distribución de una variable numérica dividiéndola en bins y contando el número de observaciones en cada bin.
Ejemplo: Visualizar la distribución de edades en una población.   


5. Boxplot `(boxplot)`:
Uso: Proporciona un resumen gráfico de una o varias variables numéricas a través de sus cuartiles.
Ejemplo: Comparar las puntuaciones de varios grupos de estudiantes.


6. Violin plot `(violinplot)`:
Uso: Similar al boxplot, pero también muestra la distribución de la data en diferentes niveles.
Ejemplo: Comparar la distribución salarial entre diferentes departamentos de una empresa.


7. Pair plot `(pairplot)`:
Uso: Muestra gráficos de dispersión entre pares de variables y histogramas para variables individuales, ideal para explorar correlaciones.
Ejemplo: Análisis inicial de un conjunto de datos con múltiples variables numéricas.  


8. Heatmap `(heatmap)`:
Uso: Representa datos en una matriz como valores de colores, muy útil para representar correlaciones o matrices de confusión.
Ejemplo: Mostrar la correlación entre diferentes variables en un conjunto de datos. (se realiza con Seaborn)


9. Swarm plot `(swarmplot)`:
Uso: Representa los puntos exactos de datos y puede ser utilizado junto con boxplots o violin plots para mostrar la distribución de los datos.
Ejemplo: Mostrar las puntuaciones de varios grupos de estudiantes, mostrando cada puntuación individualmente. (se realiza con Seaborn)

## Buenas prácticas al graficar

**Etiquetas claras**
- Título: Siempre incluye un título descriptivo.
- Ejes: Etiqueta los ejes x e y para indicar qué variable se está mostrando.
- Leyenda: Si hay múltiples trazados, incluye una leyenda

```python
## plt.title("Evolución de las ventas en el tiempo")
## plt.xlabel("Tiempo (meses)")
## plt.ylabel("Ventas ($)")
## plt.legend()
```

**Escalas Apropiadas:** 
- Asegúrate de que las escalas de los ejes sean apropiadas para los datos. No es recomendable tener espacios vacíos significativos o truncar datos.

**Colores:**
- Usa colores que se distingan entre sí, especialmente si hay múltiples trazados.
- Ten en cuenta a las personas con daltonismo; hay paletas de colores diseñadas específicamente para ser accesibles.

**Tamaño de Fuente:**
- Asegúrate de que todas las etiquetas y títulos sean legibles, incluso cuando la gráfica se vea en un tamaño reducido.

**Evita Gráficos Engañosos:**
- No manipules las escalas para exagerar diferencias.
- Si no empiezas un gráfico de barras desde cero, es esencial indicarlo claramente, ya que puede ser engañoso.

**Simplicidad:** 
- No sobrecargues la gráfica con demasiada información. Si una gráfica se siente abrumadora, considera dividirla en varias más simples.

**Rejillas (Grids)**: 
- En algunos casos, un grid suave puede ayudar a leer valores con precisión.

```python
## plt.grid(True, which="both", linestyle="--", linewidth=0.5)
```

**Consistencia:** 
- Si estás mostrando varias gráficas juntas (por ejemplo, en una presentación o informe), intenta mantener un estilo y diseño consistentes entre ellas.

## Pie chart: Componentes del PBI

```python
# Aislando un solo vector para el 2021
row = df[df['Año'] == 2021].iloc[0]

data = [row['PBI Sectores Primarios'], row['PBI Sectores no Primarios']]
labels = ['PBI Sectores Primarios', 'PBI Sectores no Primarios']
colors = ['blue', 'red']
explode = (0.1, 0)  # Explotar para visibilidad

plt.figure(figsize=(8, 8))
plt.pie(data, explode=explode, labels=labels, colors=colors,
        autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '',
        textprops={'color': "white"}, shadow=True, startangle=140)

plt.title("Año 2021")

plt.tight_layout()
plt.show()
```

```python
years = [2020, 2021]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 8))

colors = ['blue', 'red']
explode = (0.1, 0)  # Explotar para visibilidad

for ax, year in zip(axes, years):
    row = df[df['Año'] == year].iloc[0]
    data = [row['PBI Sectores Primarios'], row['PBI Sectores no Primarios']]
    labels = ['PBI Sectores Primarios', 'PBI Sectores no Primarios']
    
    ax.pie(data, explode=explode, colors=colors,
           autopct=lambda p: '{:.1f}%'.format(p) if p > 0 else '',
           textprops={'color': "white"}, shadow=True, startangle=140)
    
    ax.set_title(f"Año {year}")
    
plt.legend(labels, frameon=False, loc='upper center', ncol=2, bbox_to_anchor=(-0.2, 0.0))
#plt.tight_layout()
plt.show()
```

```python
filtered_df = df[df['Año'].isin(range(2015, 2020))]

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(10, 15))

# Loop through each year and plot
for i, (index, row) in enumerate(filtered_df.iterrows()):
    data = [row['PBI Sectores Primarios'], row['PBI Sectores no Primarios']]
    labels = ['PBI Sectores Primarios', 'PBI Sectores no Primarios']
    colors = ['blue', 'green']
    explode = (0.1, 0)  # Explotar el primer triangulo 
    
    axes[i].pie(data, explode=explode, labels=labels, colors=colors,
                autopct=lambda p: '{:.1f}%'.format(p).format(p) if p > 0 else '',  # Using lambda for formatting
                textprops={'color': "white"},  # color blanco para el tetxoto
                shadow=True, startangle=140)
    
    axes[i].set_title(f"Año {row['Año'].astype(int)}")

plt.tight_layout()
plt.show()
```
